```python
#Day 4: Exploratory Data Analysis on Iris Dataset

# 1. Εισαγωγή βιβλιοθηκών
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Φόρτωμα Iris dataset (από seaborn)
df = sns.load_dataset("iris")

# 3. Ματιά στα δεδομένα
print("----- 5 πρώτες γραμμές -----")
print(df.head())

print("\n----- Πληροφορίες -----")
print(df.info())

print("\n----- Στατιστικά -----")
print(df.describe())

# 4. Έλεγχος μοναδικών κατηγοριών
print("\nΚατηγορίες είδους λουλουδιού:", df["species"].unique())

# 5. Οπτικοποιήσεις
# Κατανομή μήκους πετάλου (petal_length)
plt.figure(figsize=(6,4))
sns.histplot(df["petal_length"], bins=20, kde=True, color="purple")
plt.title("Κατανομή μήκους πετάλου")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")
plt.show()

# Μέσο μήκος/πλάτος πετάλου ανά είδος
plt.figure(figsize=(6,4))
sns.barplot(x="species", y="petal_length", data=df, estimator="mean", ci=None, palette="Set2")
plt.title("Μέσο μήκος πετάλου ανά είδος")
plt.show()

# Σχέση sepal_length - sepal_width
plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df, palette="Set1")
plt.title("Sepal Length vs Sepal Width")
plt.show()

# Pairplot για όλες τις μεταβλητές
sns.pairplot(df, hue="species", diag_kind="kde", palette="husl")
plt.suptitle("Σχέσεις μεταβλητών του Iris Dataset", y=1.02)
plt.show()
```
